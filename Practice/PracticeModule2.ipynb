### Loops


```python
#For Loop
for i in range(10):
    print(i, end=" ")
```

```python
for i in range(1,21,2):
    print(i, end=' ')
```

```python
#even or odd
lst = [1,4,72,2,3,23,11,90,88,45,14,67]
for index, number in enumerate(lst):
    print(f'Index: {index}, number:{number}')
```

```python
for i in lst:
    if i%2==0:
        print(f"{i} is even")
    else:
        print(f"{i} is odd")
```

```python
#Sort the list
sorted_lst = sorted(lst)
print(sorted_lst)
```

```python
#reverse list
reversed_lst = sorted_lst[::-1]
print(reversed_lst)
```

```python
for i in range(10,0,-2):
    print(i, end=' ')
```

```python
string = "Hello World!"
for ch in string:
    print(ch, end=' ')

#String slicing
print('')
print(string[:3])
print(string[1:4])
print(string[9:])
print(string[-1])
```

### while loop

```python
count=0
while count<=5:
    print(count, end=' ')
    count+=1
```

```python
while True:
    if count == 10:
        break  #Break statement
    print(count, end=' ')
    count+=1
```

```python
#continue statement
for i in range(10):
    if i%2==0:
        continue
    print(i)
```

```python
for i in range(1,11):
    if i==5:
        print(f'The number is: {i}')
        pass
    print(i)
```

```python
#Nested Loop
for i in range(3):
    for j in range(4):
        print(f'({i}, {j}) ', end=" ")
    print()
```

```python
#sum of first n natural number
n=10
i=0
sum_of_numbers=0
while i<=n:
    sum_of_numbers +=  i
    i+=1

print(sum_of_numbers)
```
