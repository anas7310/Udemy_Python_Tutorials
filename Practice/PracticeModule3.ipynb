```python
squared_numbers=[]
for i in range(1,11):
    squared_numbers.append(i**2)

print(squared_numbers)
```

```python
#list comprehension
lst=[x for x in range(1,11)]
print(lst)
```

```python
squared_nums = [x**2 for x in range(1,15)]
print(squared_nums)
```

```python
even_numbers = [x for x in range(1,31) if x%2==0 and x>0]
```

```python
even_numbers
```

```python
#Nested list comprehension
lst1=[1,2,3,4,5]
lst2=['a','b','c','d','e','f']

pair = [[x,y] for x in lst1 for y in lst2]
print(pair)
```

```python
words = ['hello','world','python','list','comprehension']
lengths = [len(word) for word in words]
print(lengths)
```

```python
numbers = (1,2,3,4,6,7,8,9,1,1,2,4,63,2,23,3,15)
print(numbers)
```

```python
print(numbers.index(6)) #find the index of number 6
print(numbers.count(1)) #find the occurrence of number 1
```

```python
packed_tuple = (1,True,'Hello',3.14)
print(packed_tuple)
```

```python
a,b,c,d = packed_tuple
print(a)
print(b)
print(c)
print(d)
```

```python
#unpacking with *
first, *middle,last = numbers
print(first)
print(middle)
print(last)
```
