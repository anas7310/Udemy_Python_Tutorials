## Diabetes Prediction using Decision Tree

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from MachineLearning.Practical.MultipleLinearRegressionProject import regressor

%matplotlib inline
```

```python
from sklearn.datasets import load_diabetes
```

```python
dataset  =load_diabetes()
```

```python
dataset
```

```python
X = pd.DataFrame(dataset['data'], columns=['age','sex','bmi','bp','s1','s2','s3','s4','s5','s6'])
```

```python
X.head()
```

```python
y = dataset['target']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.30, random_state=10)
```

```python
X_train.head()
```

```python
#Correlation
X_train.corr()
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(X_train.corr(), cmap='crest', annot=True)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor()
regressor.fit(X_train, y_train)
```

## Hyper Parameter Tuning

```python
param = {
    'criterion':['squared_error', 'friedman_mse', 'absolute_error', 'poisson'],
    'splitter':['best', 'random'],
    'max_depth':[1,2,3,4,5,10,15,20,25],
    'max_features':['auto','sqrt','log2']
}
```

```python
param
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
regressor = DecisionTreeRegressor()
```

```python
grid = GridSearchCV(regressor, param_grid=param, cv = 5,  scoring='neg_mean_squared_error')
```

```python
import warnings
warnings.filterwarnings('ignore')
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
y_pred = grid.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
print(r2_score(y_test, y_pred))
print(mean_squared_error(y_test, y_pred))
print(mean_absolute_error(y_test, y_pred))
```

```python
selectedModel = DecisionTreeRegressor(criterion='poisson', max_depth=4, max_features='log2', splitter='best')
```

```python
selectedModel.fit(X_train, y_train)
```

```python
##Vizualise the decision tree
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(selectedModel, filled=True)
```

```python
y_selected_pred = selectedModel.predict(X_test)
```

```python
print(r2_score(y_test, y_selected_pred))
print(mean_squared_error(y_test, y_selected_pred))
print(mean_absolute_error(y_test, y_selected_pred))
```
