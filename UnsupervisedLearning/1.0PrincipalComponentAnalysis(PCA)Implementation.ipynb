```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
#load the dataset
from sklearn.datasets import load_breast_cancer
```

```python
cancer_dataset = load_breast_cancer()
```

```python
cancer_dataset.keys()
```

```python
print(cancer_dataset.DESCR)
```

```python
df = pd.DataFrame(cancer_dataset['data'], columns=cancer_dataset['feature_names'])
```

```python
df.head()
```

# Dimensionality Reduction, not a classification or regression problem

```python
##Standardization
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaler.fit(df)
```

```python
scaled_dataset = scaler.transform(df)
```

```python
scaled_dataset
```

```python
## Applying PCA Algorithms
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=2)
```

```python
data_pca=pca.fit_transform(scaled_dataset)
```

```python
data_pca
```

```python
pca.explained_variance_
```

```python
plt.figure(figsize=(8,6))
plt.scatter(data_pca[:,0], data_pca[:,1], c=cancer_dataset['target'], cmap='plasma')
plt.xlabel("First Principal component")
plt.ylabel("Second Principal component")
```
