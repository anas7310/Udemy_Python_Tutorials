```python
#Filling missing values with the mean of the column
df['product_star_rating_filled'] = df['product_star_rating'].fillna(df['product_star_rating'].mean())
```

```python
#Renaming columns
df = df.rename(columns={'product_star_rating':'product_rating'})
df.head()
```

```python
#Change datatypes
df['product_rating_new'] = df['product_rating'].fillna(df['product_rating'].mean()).astype(int)
df
```

```python
df['Rating_Lambda'] = df['product_rating_new'].apply(lambda x: x**2)
df
```

```python
df = pd.read_csv('sample_data.csv')
```

```python
df
```

```python
df.isnull().any()
```

```python
df['Value_Filled'] = df['Value'].fillna(df['Value'].mean())
```

```python
df
```

```python
df.dtypes
```

```python
grouped_mean = df.groupby('Product')['Value_Filled'].mean()
```

```python
grouped_mean
```

```python
grouped_sum = df.groupby(['Product','Region'])['Value_Filled'].sum()
grouped_sum
```

```python
#aggregate multiple functions
df['Value'].fillna(df['Value'].mean())
grouped_agg = df.groupby('Region')['Value'].agg(['mean','sum','count'])
```

```python
grouped_agg
```

```python
##merging and joining dataframes
#Create sample DataFrames
df1 = pd.DataFrame({'Key':['A','B','C'], 'Value1':[1,2,3]})
df2 = pd.DataFrame({'Key':['A','B','D'], 'Value1':[4,5,6]})
```

```python
df1
```

```python
df2
```

```python
#Merge Dataframes on the "Key" columns
pd.merge(df1, df2, on='Key', how='inner')
```

```python
pd.merge(df1, df2, on='Key', how='outer')
```
